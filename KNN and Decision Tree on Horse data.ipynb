```python
f= f.drop(['abdomo_protein', 'abdomo_appearance','abdomen','nasogastric_reflux_ph','nasogastric_tube','nasogastric_reflux', 'rectal_exam_feces'], axis=1)
```

```python
f.isnull().sum()
```

```python
# get all categorical columns in the dataframe
catCols = [col for col in f.columns if f[col].dtype=="O"]

from sklearn.preprocessing import LabelEncoder

lb_make = LabelEncoder()

for item in catCols:
    f[item] = lb_make.fit_transform(f[item])
```

```python
f.describe()
```

```python
f.fillna(f.mean(), inplace = True)
f
```

```python
f.isnull().sum()
```

```python
from scipy import stats
import numpy as np
```

```python
z_scores = stats.zscore(f)
#calculate z-scores of `df`

abs_z_scores = np.abs(z_scores)
filtered_entries = (abs_z_scores < 3).all(axis=1)
new_f = f[filtered_entries]

print(new_f)
```

```python
from sklearn.model_selection import train_test_split
```

```python
new_f.columns
```

```python
x= new_f[["surgery","age","respiratory_rate","temp_of_extremities","capillary_refill_time","pain","abdominal_distention","packed_cell_volume","total_protein"]]
```

```python
y = new_f["outcome"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=4) 
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors = 1)
```

```python
knn.fit(X_train, y_train)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print( confusion_matrix(y_test, pred))
```

```python
print(classification_report(y_test,pred))
```

```python
np.mean(pred!=y_test)
```

```python
error_rates =[]

for i in range(1,40):
    knn = KNeighborsClassifier(n_neighbors = i)
    knn.fit(X_train, y_train)
    pred = knn.predict(X_test)
    error_rates.append( np.mean(pred!=y_test) )
```

```python
error_rates
```

```python
plt.figure(figsize=(10,6))
plt.plot(range(1,40),error_rates, color='blue', linestyle='dashed',marker='o')
#we will use k =30 t0 get the lowest error of the model./
```
